# Data Preparations

```python
import pandas as pd

# The original lending club dataset is downloaded from: https://www.kaggle.com/datasets/wordsforthewise/lending-club/data

# Read loan dataset from the folder
file_path = '/content/drive/MyDrive/Study/NLP for Finance/data/accepted_2007_to_2018Q4.csv'
df = pd.read_csv(file_path)

# Show column names which are the features of each loan application
print("--- Column Names ---")
print(df.columns.tolist())
```

```python
# Define columns to select
columns_to_select = [
    'loan_amnt',
    'term',
    'int_rate',
    'installment',
    'grade',
    'sub_grade',
    'emp_title',
    'emp_length',
    'home_ownership',
    'annual_inc',
    'verification_status',
    'issue_d',
    'purpose',
    'dti',
    'earliest_cr_line',
    'open_acc',
    'pub_rec',
    'revol_bal',
    'total_acc',
    'initial_list_status',
    'application_type',
    'mort_acc',
    'pub_rec_bankruptcies',
    'zip_code',
    'loan_status'
]

# Create new DataFrame with selected columns
selected_df = df[columns_to_select]

# Remove the rows that loan_amnt is empty
selected_df = selected_df.dropna(subset=['loan_amnt'])
```

```python
# Create mapping dictionary that rename the features to natural language
column_mapping = {
    'loan_amnt': 'Loan amount',
    'term': 'loan term',
    'int_rate': 'interest rate',
    'installment': 'monthly payment',
    'grade': 'loan grade',
    'sub_grade': 'loan subgrade',
    'emp_title': 'job title',
    'emp_length': 'years employed',
    'home_ownership': 'home ownership status',
    'annual_inc': 'annual income',
    'verification_status': 'income verification status',
    'issue_d': 'loan application issue date',
    'purpose': 'loan purpose',
    'dti': 'debt-to-income ratio',
    'earliest_cr_line': 'first credit line date',
    'open_acc': 'number of open credit accounts',
    'pub_rec': 'derogatory public records',
    'revol_bal': 'revolving balance',
    'total_acc': 'total credit accounts',
    'initial_list_status': 'initial listing status',
    'application_type': 'application type',
    'mort_acc': 'mortgage accounts',
    'pub_rec_bankruptcies': 'bankruptcy records',
    'zip_code': 'ZIP code',
    'chargeoff_within_12_mths': 'Charge-offs',
    'loan_status': 'loan status'
}

# Rename the columns
selected_df = selected_df.rename(columns=column_mapping)
```

```python
import numpy as np
np.random.seed(42)

# Select "fully paid" loans as "good" loans
fully_paid_df = selected_df[selected_df['loan status'] == 'Fully Paid']

# Drop the loan status column as it will be training target later
fully_paid_df.drop('loan status', axis=1, inplace=True)

# Randomly select 316,824 data points to be in line with baseline
fully_paid_df = fully_paid_df.sample(n=316824, random_state=42)

print(fully_paid_df.shape)
```

```python
# First create the text description for each row
def create_sentence(row):
    sentence_parts = []
    for column in row.index:
        sentence_parts.append(f"{column} is {row[column]}")
    return ", ".join(sentence_parts)

# Apply the function to each row and create new DataFrame
text_descriptions = fully_paid_df.apply(create_sentence, axis=1)

# Create new DataFrame with text and label columns
good_df = pd.DataFrame({
    'text': text_descriptions,
    'label': 'good'
})
```

```python
# Select the loans met the requirements as "bad" loans
default_df = selected_df[selected_df['loan status'].isin([
    'Charged Off',
    'Default',
    'Late (31-120 days)',
    'Does not meet the credit policy. Status:Charged Off'
])]

# Drop the loan status column as it will be training target later
default_df.drop('loan status', axis=1, inplace=True)

# Randomly select 79,206 data points to be in line with baseline
default_df = default_df.sample(n=79206, random_state=42)

print(default_df.shape)
```

```python
# Apply the function to each row and create new DataFrame
text_descriptions_bad = default_df.apply(create_sentence, axis=1)

# Create new DataFrame with text and label columns
bad_df = pd.DataFrame({
    'text': text_descriptions_bad,
    'label': 'bad'
})
bad_df
```

```python
# Concatenate the two dataframes vertically
combined_df = pd.concat([good_df, bad_df], axis=0)

# Shuffle all rows randomly
# Reset index after shuffling to have clean continuous indices
shuffled_df = combined_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Verify the shape
print(f"Shape of combined and shuffled DataFrame: {shuffled_df.shape}")
```

# Fine-tuning Llama-3.2-1B via QLoRA

```python
import torch
from transformers import (
    AutoModelForSequenceClassification,
    AutoTokenizer,
    TrainingArguments,
    Trainer,
    BitsAndBytesConfig
)
from peft import (
    LoraConfig,
    PeftConfig,
    PeftModel,
    get_peft_model,
    prepare_model_for_kbit_training
)
import pandas as pd
from datasets import Dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, f1_score
from accelerate import Accelerator
import numpy as np
import bitsandbytes as bnb

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    # Calculate F1 score
    f1 = f1_score(labels, predictions)

    # Calculate probabilities for AUC
    probs = torch.nn.functional.softmax(torch.tensor(logits), dim=-1)
    auc = roc_auc_score(labels, probs[:, 1].numpy())

    return {
        'f1_score': f1,
        'auc_score': auc
    }

def preprocess_function(examples, tokenizer, max_length=512):
    # Convert texts to strings (in case they aren't already)
    texts = [str(text) for text in examples["text"]]

    # Tokenize the texts
    tokenized = tokenizer(
        texts,
        padding="max_length",
        truncation=True,
        max_length=max_length,
        return_tensors=None
    )

    # Add labels and ensure they are integers
    tokenized['labels'] = [int(label) for label in examples['label']]

    return tokenized

def prepare_data(df):
    # Ensure labels are integers
    df['label'] = df['label'].astype(int)

    # Split the data
    train_df, eval_df = train_test_split(df, test_size=0.3, random_state=42)

    # Convert to HuggingFace datasets
    train_dataset = Dataset.from_pandas(train_df)
    eval_dataset = Dataset.from_pandas(eval_df)

    return train_dataset, eval_dataset

def setup_qlora_fine_tuning(model_id, df, output_dir="qlora_fine_tuned_llama"):
    # Initialize accelerator
    accelerator = Accelerator()

    # Load and configure tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "right"

    # Load model with quantization and padding token configuration
    quantization_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4"
    )

    model = AutoModelForSequenceClassification.from_pretrained(
        model_id,
        num_labels=2,
        torch_dtype=torch.float16,
        device_map="auto",
        quantization_config=quantization_config,
        pad_token_id=tokenizer.pad_token_id  # Add this line
    )

    # Set pad token in model config
    model.config.pad_token_id = tokenizer.pad_token_id
    model.config.use_cache = False  # Add this for training

    # Configure LoRA
    lora_config = LoraConfig(
        r=16,
        lora_alpha=32,
        target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
        lora_dropout=0.05,
        bias="none",
        task_type="SEQ_CLS"
    )

    # Get PEFT model
    model = get_peft_model(model, lora_config)

    # Prepare datasets
    train_dataset, eval_dataset = prepare_data(df)

    # Tokenize datasets
    tokenized_train = train_dataset.map(
        lambda x: preprocess_function(x, tokenizer),
        batched=True,
        remove_columns=train_dataset.column_names
    )
    tokenized_eval = eval_dataset.map(
        lambda x: preprocess_function(x, tokenizer),
        batched=True,
        remove_columns=eval_dataset.column_names
    )

    # Training arguments
    training_args = TrainingArguments(
        output_dir=output_dir,
        learning_rate=2e-4,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        num_train_epochs=1,
        weight_decay=0.01,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        push_to_hub=False,
        gradient_accumulation_steps=2,
        fp16=False,  # Changed from True to False
        bf16=True,   # Added bf16 instead
        logging_steps=10,
        optim="paged_adamw_32bit",
        report_to=["none"]
    )

    # Create Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_eval,
        tokenizer=tokenizer,
        compute_metrics=compute_metrics
    )

    return trainer, model, tokenizer

def train_model(df, model_id="meta-llama/Llama-3.2-1B"):
    # Split into train, val, and test
    train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)
    val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

    # Train the model
    trainer, model, tokenizer = setup_qlora_fine_tuning(
        model_id,
        pd.concat([train_df, val_df])
    )

    print("Starting training...")
    trainer.train()

    # Prepare test dataset
    test_dataset = Dataset.from_pandas(test_df)
    tokenized_test = test_dataset.map(
        lambda x: preprocess_function(x, tokenizer),
        batched=True,
        remove_columns=test_dataset.column_names
    )

    # Evaluate on test set
    print("\nEvaluating on test set...")
    test_results = trainer.evaluate(tokenized_test)

    print("\nTest Set Results:")
    print(f"F1 Score: {test_results['eval_f1_score']:.4f}")
    print(f"AUC Score: {test_results['eval_auc_score']:.4f}")

    # Save model
    peft_model_path = f"{trainer.args.output_dir}/final"
    trainer.model.save_pretrained(peft_model_path)
    tokenizer.save_pretrained(peft_model_path)

    return trainer, model, tokenizer, test_results

# Usage:
"""
trainer, model, tokenizer, test_results = train_model(shuffled_df)

# Print final metrics
print("\nFinal Test Set Results:")
print(f"F1 Score: {test_results['eval_f1_score']:.4f}")
print(f"AUC Score: {test_results['eval_auc_score']:.4f}")
"""
```
